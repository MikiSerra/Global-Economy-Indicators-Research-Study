La variable que más nulos tiene es la variación de inventarios, con un 17% de nulos, seguida de las actividades de la agricultura, con un 1%, e inferiores en otros casos.

El origen de estos nulos puede deberse a gran variedad de casuísticas como errores de registro, inexistencia del país en ciertos momentos de la historia y hasta la no elaboración de la estadística *per se*. Además, el tratamiento de los nulos en el ámbito de las Series Temporales merece un tratamiento y reflexión diferentes al de un DataFrame. No podemos sustituir por la media, moda o mediana pues estas variables no contemplan la dimensión del tiempo e incurriríamos en errores debido a la naturaleza y tendencia alcista de nuestras variables (por ejemplo, adjudicarle a un hipotético PIB nulo de España en 1983 el valor del PIB promedio de ese país con el resto de años, generando un pico anómalo en nuestra serie de tiempo).

Por ello, y debido a que la cantidad de nulos tampoco es preocupante (1% o menor), decidimos no dropearlos, pero podría quedar como ejercicio el intentar predecirlos con Machine Learning mediante modelos de Regresión o con modelos de Series Temporales

### Continente

```python
# Definimos una función para obtener el continente de un país
def country_to_continent(country_name):
    """
    Creamos una función con la que podamos obtener el continente de un país. Para ello, utilizaremos la librería pycountry_convert.

    Args:
    country_name: str - Nombre del país

    Returns:
    continent_name: str - Nombre del continente
    """
    try:
        # Obtenemos el código del país
        country_alpha2 = pc.country_name_to_country_alpha2(country_name)

        # Obtenemos el código de continente a partir del código de país
        continent_code = pc.country_alpha2_to_continent_code(country_alpha2)

        # Convertimos el código de continente a nombre de continente
        continent_name = pc.convert_continent_code_to_continent_name(continent_code)
        return continent_name
    except:
        # Devolver 'Unknown' para excepciones, es decir, cuando no se puede obtener el código el país 
        return "Unknown"

# Aplicamos la función para crear una nueva columna de continentes 
df['Continent'] = df['Country'].apply(country_to_continent)
```

```python
df["Continent"].value_counts()
```

Vemos que hay algunos países de los que no se ha podido obtener su código y ha devuelto como valor de continente un Unknown. Para arreglar esto, vamos a ver qué países se han quedado fuera de la función y vamos a colocarlos manualmente

```python
unknown = df[df["Continent"] == "Unknown"]["Country"].unique()
unknown
```

```python
Africa = ['D.R. of the Congo', 'Ethiopia (Former)', 'Sudan (Former)', 'U.R. of Tanzania: Mainland', 'Zanzibar']
Asia = ['State of Palestine', 'China, Hong Kong SAR', 'Iran (Islamic Republic of)','USSR (Former)', 'Yemen Arab Republic (Former)', 
        'D.P.R. of Korea', 'Republic of Korea', "Lao People's DR", 'China, Macao SAR', 'Micronesia (FS of)', 'Yemen Democratic (Former)', 'Timor-Leste'] 
Europe =['Czechoslovakia (Former)', 'Yugoslavia (Former)', 'Kosovo', 'Former Netherlands Antilles', 'Sint Maarten (Dutch part)']
South_America =['Bolivia (Plurinational State of)', 'St. Vincent and the Grenadines', 'Venezuela (Bolivarian Republic of)']

df['Continent'] = df.apply(
    lambda row: 'Africa' if row['Continent'] == 'Unknown' and row['Country'] in Africa else
                'Asia' if row['Continent'] == 'Unknown' and row['Country'] in Asia else
                'Europe' if row['Continent'] == 'Unknown' and row['Country'] in Europe else
                'South America' if row['Continent'] == 'Unknown' and row['Country'] in South_America else
                row['Continent'],
    axis=1
)
```

### PIB per cápita

Una medida para *intentar* representar la riqueza de la población para permitir comparaciones.

```python
df["GDP_pc"] = df["GDP"]/(df["Population"])
```

### Tasa de crecimiento del PIB

Calculada como la tasa de variación del PIB respecto al año anterior.

```python
df["GDP_growth"] = df.groupby("Country")["GDP"].pct_change()*100
```

### Saldo Neto Externo

También conocida como la Balanza Comercial, representa el superávit o déficit comercial con el sector exterior.

```python
df["SNE"] = df["Exports"] - df["Imports"] 
```

### Índice de Apertura Comercial

Cálculo del peso conjunto de las exportaciones y las importaciones sobre el total de la economía del país. Indica el peso o importancia que supone el sector exterior en dicho país.

```python
df["IAC"] = ((df["Exports"] + df["Imports"])/df["GDP"])*100 
```

### Tasa de cobertura comercial

Cálculo del peso que suponen las exportaciones sobre el total de importaciones. Si la ratio es mayor que 1 (>100%), significa que las exportaciones son superiores a las importaciones (SNE>0). Sirve para relativizar un poco el SNE y para poder comparar.

```python
df["TCC"] = df["Exports"]/df["Imports"]*100
```

### Exportaciones sobre PIB

Cálculo del peso que suponen las exportaciones sobre el total de la economía. Ayuda a tener una idea de la dependencia que tiene el país de sus exportaciones al extranjero

```python
df["Exp/GDP"] = (df["Exports"]/df["GDP"])*100 
```

### Convergencia dentro de la EuroZona?

Definiremos la Eurozona como el conjunto de países cuya divisa oficial sea el euro.

```python
Eurozona = df[df["Currency"] == "Euro"].drop(columns=["Currency"])
```

```python
Eurozona["Country"].unique()
```

```python
# Creamos una variable en el dataframe Eurozona que indique el GDPpc promedio de todos los paises en cada año
Eurozona["GDPpc_Avg"] = Eurozona.groupby("Year")["GDP_pc"].transform("mean")

# Y una variable que indique la ratio entre el GDPpc de cada país cada año sobre el GDPpc promedio de todos los paises en ese año
Eurozona["GDPpc_Convergency_Ratio"] = (Eurozona["GDP_pc"]/Eurozona["GDPpc_Avg"])*100
```

#### ¿Y si partimos más el tramo de PIBpc?

```python
# Segmentamos en 3 tramos de PIBpc
df2=df.copy()
df2=df2[df2["Year"] == 2021]
df2['GDP_category'] = df2['GDP_pc'].apply(lambda x: 'High' if x > 10000 else('Low' if x < 3000 else 'Medium'))

# Separamos los datos en 3 grupos en funcion del GDP_pc
idh_high = df2[df2['GDP_category'] == 'High']['IDH'].dropna()
idh_medium = df2[df2['GDP_category'] == 'Medium']['IDH'].dropna()
idh_low = df2[df2['GDP_category'] == 'Low']['IDH'].dropna()
unemployment_high = df2[df2['GDP_category'] == 'High']['Unemployment'].dropna()
unemployment_medium = df2[df2['GDP_category'] == 'Medium']['Unemployment'].dropna()
unemployment_low = df2[df2['GDP_category'] == 'Low']['Unemployment'].dropna()

# Test ANOVA
anova_idh = stats.f_oneway(idh_high, idh_medium, idh_low)
anova_unemployment = stats.f_oneway(unemployment_high, unemployment_medium, unemployment_low)

print(f"ANOVA para IDH: estadístico F = {anova_idh.statistic}, valor p = {anova_idh.pvalue}")

if anova_idh.pvalue < 0.05:
    print('Rechazamos la hipótesis nula de que el IDH promedio en 2021 es igual entre los grupos, por lo que es diferente entre, al menos 2 de los grupos de países definidos por tramos de PIB pc')
else:
    print('No Rechazamos la hipótesis nula de que el IDH promedio en 2021 es igual entre los grupos de países definidos por tramos de PIB pc')

print()
print(f"ANOVA para Desempleo: estadístico F = {anova_unemployment.statistic}, valor p = {anova_unemployment.pvalue}")

if anova_unemployment.pvalue < 0.05:
    print('Rechazamos la hipótesis nula de que la tasa de desempleo promedio en 2021 es igual entre los grupos, por lo que es diferente entre, al menos 2 de los grupos de países definidos por tramos de PIB pc ')
else:
    print('No Rechazamos la hipótesis nula de que la tasa de desempleo promedio en 2021 es igual entre los grupos de países definidos por tramos de PIBpc')
```

Efectivamente, si sabemos que los tramos por encima y debajo de 10k son diferentes, realizar un test ANOVA con los tramos más desglosados es muy probable que devuelva la misma conclusión, pues al menos 2 de los 3 grupos no son iguales

Con una diferenciación de orden 1 y 2, seríamos capaces de rechazar la hipótesis nula de que la serie no es estacionaria en media. Vamos a utilizar la diferenciación de orden 1 para perder la menor cantidad de datos posible

Probaremos con ARIMA(p,1,q). Utilizando un auto ARIMA se probarán automáticamente varios modelos y se seleccionará el que menor criterio de información de Akaike (AIC) tenga

```python
model = auto_arima(Eurozona['GDP'], start_p=1, start_q=1,
                      test='adf',
                      max_p=5, max_q=5,
                      m=1,             
                      d=1,          
                      seasonal=False,   
                      start_P=0, 
                      D=None, 
                      trace=True,
                      error_action='ignore',  
                      suppress_warnings=True, 
                      stepwise=True)
```

Se ha detectado automáticamente que el mejor modelo según el AIC es ARIMA(0,1,0)

Destaca el crecimiento casi constante a lo largo de toda la serie del peso de la inversión privada sobre el PIB de China, que desde 2004 ya equivalía aproximadamente al peso en la Eurozona y en EE.UU. conjuntamente.

<FONT COLOR=#FCB714> **Conclusiones Interregionales**:</FONT>
- Paro: España, Grecia y Montenegro son las regiones con más paro, cuyas tasas triplican las de, en contraposición, Luxemburgo, Austria y Países Bajos.
- Consumos públicos: Irlanda, Kosovo y Luxemburgo son los países con menor peso del estado en cuanto a lo que consumos públicos refiere mientras que en Francia, Mónaco y Países Bajos casi se duplica esta ratio.
- IDH: Casi todos los países tienen un IDH de muy alto y los que menor IDH tienen no bajan de 0,8.
- PIB y Población: Alemania, Francia, Italia, España, Países Bajos y Bélgica componen los mayores componentes económicos y poblacionales de la Zona Euro, suponiendo, todos ellos en conjunto, una porción bastante elevada de todos los sectores económicos.
- PIB per cápita: Por otro lado, Mónaco, Luxemburgo, Irlanda, San Marino, Países Bajos y Austria encabezan la lista de la riqueza media por persona. Habrá que estudiar qué tan bien se reparte la riqueza en estas regiones.
- Convergencia: Es complicado hablar de convergencia. Hay países como Mónaco, Luxemburgo y San Marino que se ven aparentemente muy por encima de la media europea, pero puede estar provocado por la poca población que hay en estos lugares, aunque también tenemos el caso de Malta y Montenegro, con apenas 1 millón de personas entre ambos y al fondo de las listas en convergencia. Excluyendo esos países más reducidos, países como Irlanda, Finlandia, Países Bajos, Austria, Bélgica y Alemania están situados por encima del 100%, cuando Francia, Italia, España, Portugal y Grecia, en la región geográfica de los denominados como "PIGS" (Portugal, Italy, Greece, Spain), están por debajo de la media. Estas diferencias se deben probablemente a un grandísimo número de casuísticas sociales, políticas, históricas, económicas, geográficas y culturales muy diversas que darían para un estudio propio. Sin embargo, los datos parecen sugerir que con los años estas diferencias se están reduciendo, muy lentamente.


<FONT COLOR=#FCB714> **Final**:</FONT><br>
<B><br>En general, la Eurozona todavía es una aglomeración de países con grandes diferencias entre sus extremos. Sería interesante plantear y estudiar la cuestión de a qué países les ha resultado útil o beneficiosa la anexión a la Unión y a qué países no les ha ido tan bien.
